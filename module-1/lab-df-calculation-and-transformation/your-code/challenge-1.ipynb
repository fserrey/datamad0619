#### Cleanup `Name` that contain "Mega".

If you have checked out the pokemon names carefully enough, you should have found there are junk texts in the pokemon names which contain "Mega". We want to clean up the pokemon names.
```python
# your code here
import re
pokemon.columns
pokemon["Name"].replace("^(.*?)Mega", "", regex = True)

# Nota para el futuro -->
# ->( )<- Todo lo que esté dentro del paréntesis hará referencia a lo que va delante de la palabra Mega.
# ->^<- Marca el principio de la string
# ->.*<- El punto y asterisco es para coger cualquier caracter, la ? es para marcar el tope hasta el que coger 


# test transformed data
#pokemon.head(10)
```

#### Create a new column called `A/D Ratio` whose value equals to `Attack` devided by `Defense`.

```python
# your code here
print(pokemon.columns)

pokemon["A/D Ratio"] = pokemon["Attack"]/pokemon["Defense"]

display(pokemon["A/D Ratio"].head())
```
